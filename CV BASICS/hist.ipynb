```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# img=cv2.imread("CVfiles/roi.jpg")
img=np.random.randint(0,255,(100,100,3),np.uint8)

plt.imshow(img)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
hist1=cv2.calcHist(img,[0],None,[16],[0,255])
hist2=cv2.calcHist(img,[1],None,[16],[0,255])
hist3=cv2.calcHist(img,[2],None,[16],[0,255])
plt.plot(hist1,"RED"),plt.plot(hist2,"GREEN"),plt.plot(hist3,"blue")
```

```python
img=cv2.imread("CVfiles/sample1.jpg",0)

hist,bins=np.histogram(img.ravel(),256,[0,256])
# print(len(hist))


cdf=hist.cumsum()
print(cdf)
print(len(cdf))


# fig, ax = plt.subplots(3)

cdf_normalized = cdf * float(hist.max()) / cdf.max()


# ax[0].plot(cdf_normalized, color = 'b')
# ax[1].hist(img.flatten(),256,[0,256], color = 'r')

plt.plot(cdf_normalized, color = 'b')
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```
