```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

```

```python
org_img=cv2.resize(cv2.imread("CVfiles/objsep/Contoso7.jpeg"),(500,500))

org_img=cv2.cvtColor(org_img,cv2.COLOR_BGR2RGB)
plt.imshow(org_img)
print(org_img.shape)
print(org_img.dtype)
```

```python
gry_img=cv2.cvtColor(org_img,cv2.COLOR_RGB2GRAY)
plt.imshow(gry_img,cmap="gray")
```

```python
thresh=cv2.threshold(gry_img,127,255,cv2.THRESH_BINARY)[1]
print(thresh.shape)
plt.imshow(thresh,cmap="gray")
```

```python
hsv=cv2.cvtColor(org_img,cv2.COLOR_RGB2HSV)
h,s,v=cv2.split(hsv)
thresh=cv2.threshold(h,77,255,cv2.THRESH_BINARY)[1]
# plt.imshow(h)
# plt.imshow(s)
# v=cv2.inRange(v,120,250)
plt.imshow(thresh,cmap="gray")
# plt.imshow(s,cmap='gray')
# plt.imshow(h,cmap='gray')
```

```python



blur=cv2.GaussianBlur(h,(3,3),0)
# plt.imshow(h)
hist=cv2.calcHist(h,[0],None,[180],[0,180])
# hist=cv2.calcHist(h,[0],None,[180],[0,180])

print(np.unique(hist))
# plt.plot(hist)
plt.imshow(h)
print(h.shape)
```

```python
import random

cnt,hierarchy=cv2.findContours(h,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
# print(cnt)

for contour in cnt[1:]:
    
    cont=contour.ravel()
    
    color = random.sample(range(255),3)
    print(color)

    for j in range(1,len(cont),2):
        cv2.circle(org_img,(cont[j-1],cont[j]),10,color,1)
plt.imshow(org_img)
```
